__2.__ Consider values $N$ = 1,000 and $N$ = 10,000. There are $M/N$ samples of this size in our $M$ values. Histogram the sample means for these values of $N$ and determine the true standard deviation of the means $\hat{\sigma}_{\bar{v}_{a},N}$.

Recall that $\hat{\sigma}_{\bar{v}_{a},N} = \frac{\sigma}{\sqrt{N}}$, where $\sigma$ is the variance of the overall population. So we should see that $\bigg(\frac{\hat{\sigma}_{\bar{v}_{a},N_{1}}}{\hat{\sigma}_{\bar{v}_{a},N_{2}}}\bigg)^{2} = \frac{N_{2}}{N_{2}} = 10$, in this case. This is roughly the result we get (see below).

```julia
include("q1/p2.jl")
```

```julia
# Histogram for dataset 1
using Gadfly
p2 = plot(
    layer(
        x = v̄1[:,1],
        color = ["N1 = 1,000" for i in 1:25],
        Geom.histogram(bincount=25, density=true),
        order=1
    ),
    layer(
        x = v̄2[:,1],
        color = ["N2 = 10,000" for i in 1:25],
        Geom.histogram(bincount=25, density=true), 
        order=2,
        Theme(default_color=color("red"))
    ),
    Guide.xlabel("Sample mean"),
    Guide.ylabel("Relative Frequency"),
    Guide.title("Sample Mean Histogram in Dataset 1"),
    Guide.colorkey("Legend title"),
)
```
